```python
import matplotlib.pyplot  as plt
```

```python
# plot distribution of power
plt.figure(figsize=(10, 6))
sns.histplot(df['generated_power_kw'],bins=30,kde=True)
plt.title('Distribution of Generated Power(KW)')
plt.xlabel('Generated power(KW)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Boxplot for detecting outliers in generated_power_kw
plt.figure(figsize=(6, 4))
sns.boxplot(x=df['generated_power_kw'], color='lightcoral')
plt.title('Boxplot of Generated Power (kW)')
plt.xlabel('Generated Power (kW)')
plt.show()
```
